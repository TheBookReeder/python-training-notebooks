### Intro to Seaborn

````python
# Initial Setup
%matplotlib inline

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

grants = pd.read_csv('grants.csv')
ins_prems = pd.read_csv('insurance_premiums.csv')
````

Pandas has it's own built in plotting function that is also built off matplotlib - should look fairly familiar, see below

````python
grants['Award_Amount'].plot.hist()
plt.show()
````

### distplot
Now, have a look at the aesthetical upgrade and a statistical overlay that seaborn offers in the default settings

````python
sns.distplot(grants['Award_Amount'])
plt.show()
````

A distplot not only gives a histogram of the data but includes a (gaussian kernel density estimate) distribution over the top. This is not always desired and can be turned off. We will also increase the number of bins of the histogram.

````python
sns.distplot(grants['Award_Amount'],
             kde=False,
             bins=20)

plt.show()
````

Since the distribution can be more relevant, we can turn off the histogram, add a rugplot (give some idea of distribution of data) and fill in the plot

````python
sns.distplot(grants['Award_Amount'],
             hist=False,
             rug=True,
             kde_kws={'shade':True})

plt.show()
````

### Countplot
Like distplot but for categorical data, and therefore doesn't include a distribution plot

````python
sns.countplot(grants[grants.State)
plt.show()
````

### Barplot
All you could want from a bar graph with a litte more done to the aesthetics

````python
x = np.array(list("ABCDEFGHIJ"))
y = np.arange(1, 11)

sns.barplot(x=x, y=y)

plt.show()
````

### Regplot
These plots could be created fairly easily using matplotlib but not as easy as seaborn makes it. Next we will look at a regression plot, which gives a line of best fit plus a confidence interval. This would not be easy to perform using ```matplotlib``` since a lot of the prep work would have to be done by the user.

**Note:** while ```distplot``` accepts a one dimensional data set (like ```pandas.Series```), which is the nature of a histogram, the ```regplot``` accepts an entire ```pandas.DataFrame``` and in the arguements, the user specifies which features of the DataFrame are assigned to the x and y axes

````python
sns.regplot(data=ins_prems,
         x="insurance_losses",
         y="premiums")

plt.show()
````

### Lmplot
```regplot``` is a scatter plot that uses a logistic regression model to fit the data - very convenient when doing an EDA or showing business the correlation between two seemingly uncorrelated features. However, it has the capability of comparing only 2 features. Another popular graph that Seaborn offers is the ```lmplot```. It uses ```regplot``` as it's backend but allows for additional 'dimensions' to be overlayed.

```lmplot``` is more computational intensive, so only use when required, otherwise stick with ```regplot```

````python
print("By Colour:")

sns.lmplot(data=ins_prems,
           x="insurance_losses",
           y="premiums",
           hue="Region")

plt.show()
````

````python
print("By Column:")

sns.lmplot(data=ins_prems,
           x="insurance_losses",
           y="premiums",
           col="Region")

plt.show()
````

### Scatter
The scatter plot without a line of best fit is just ```sns.scatterplot```, it does allow for an extra dimension: ```size```

````python
plt.figure(figsize=(10, 8))

sns.scatterplot(data=ins_prems,
                x="insurance_losses",
                y="premiums",
                hue="Region",
                size='fatal_collisions',
                sizes=(20, 300))

plt.show()
````

### Lineplot
As seen below in the first graph: the ```sns.lineplot``` is exactly the same as the ```matplotlib``` default graph. However, when multiple dependent data points exist for the same independent data point, ```seaborn``` will create a range line graph.

````python
def custom_round(x, base):
    return int(base * round(float(x)/base))

ins_prems.insurance_losses = ins_prems.insurance_losses.apply(lambda x: custom_round(x, 40))

sns.lineplot(x="insurance_losses", y="premiums",
             data=ins_prems)
plt.show()
````
