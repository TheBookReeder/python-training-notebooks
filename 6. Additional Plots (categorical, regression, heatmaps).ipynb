````python
# Initial Setup
%matplotlib inline

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

### Categorical Data (pt. 1)
We covered ```countplot``` in **4. Intro to Seaborn** which is a plot for one dimensional categorical data. However, Seaborn offers much more in-depth plots for categorical data, especially when you want to look at a 2D array of data (remember how difficult this was in matplotlib, ```meshgrid``` ring a bell?), the first of these plot we will look at is ```stripplot``` and ```swarmplot```

````python
df = pd.read_csv("schoolimprovement2010grants.csv")

fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Create the stripplot without jitter
sns.stripplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         jitter=False,
         ax=ax0)

# Create the stripplot with jitter
sns.stripplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         jitter=True,
         ax=ax1)

plt.tight_layout()
plt.show()
````

````python
plt.figure(figsize=(10, 5))

# Create and display a swarmplot with hue set to the Region
sns.swarmplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         hue='Region')

plt.show()
````

### Categorical Data (pt. 2)
These two graphs above show each and every point, but sometimes that is too much and we would rather see the "shape" of our data. The best plots for this is ```pointplots```, ```boxplot```, ```voilinplot```, and ```lvplot```

````python
# Create a pointplot and include the capsize in order to show bars on the confidence interval
sns.pointplot(data=df,
         y='Model Selected',
         x='Award_Amount',
         capsize=.1)

plt.show()
````

````python
sns.boxplot(data=df,
         x='Award_Amount',
         y='Model Selected')

plt.show()
````

````python
sns.violinplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         palette='husl')

plt.show()
````

````python
plt.figure(figsize=(10, 5))

sns.lvplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         palette='Paired',
         hue='Region')

plt.legend(loc='lower right')

plt.show()
````

### Categorical (pt. 3)
The last couple of categorical plots we will look at is: ```countplots``` and ```barplots```. We have seen these but just for completeness, these will be covered again.

````python
# Notice how we've included "hue", which is a method of most seaborn plots
sns.countplot(data=df,
         y="Model Selected",
         hue="Region")

plt.show()
````

````python
# Create a barplot with each Region shown as a different color
sns.barplot(data=df,
         y='Award_Amount',
         x='Model Selected',
         hue='Region')

plt.show()
````

### Regression Plots (pt. 1)
Linear regression is a useful tool for understanding the relationship between numerical variables. Seaborn has simple but powerful tools for examining these relationships: ```regplot``` and ```residplot```.

These have also been covered in the basics but we will look at using them more in-depth.

````python
df = pd.read_csv('college_datav3.csv')

plt.figure(figsize=(10, 8))

# Display a regression plot for Tuition
sns.regplot(data=df,
         y='Tuition',
         x="SAT_AVG_ALL",
         marker='^',
         color='g')

plt.show()
````

````python
plt.figure(figsize=(10, 8))

# Display the residual plot
sns.residplot(data=df,
          y='Tuition',
          x="SAT_AVG_ALL",
          color='g')

plt.show()
````

Residual plots are a great way of 'normalising' the dependent variable and looking for a relationship between the two variables.

### Regression Plots (pt. 2)
Let's customise the regression plot even further

````python
# Plot a regression plot of Tuition and the Percentage of Pell Grants
sns.regplot(data=df,
            y='Tuition',
            x="PCTPELL")

plt.show()
````

There is a lot of data and it is hard to see any form of relationship. We can bin the data in the ```regplot```

````python
sns.regplot(data=df,
            y='Tuition',
            x="PCTPELL",
            x_bins=5)

plt.show()
````

Much better, but this still doesn't look like a great fit - maybe a higher order regression line would be better?

````python
sns.regplot(data=df,
            y='Tuition',
            x="PCTPELL",
            x_bins=5,
            order=2)

plt.show()
````

### Heatmap
Put simply: a plot that visualises the weightings of a 2D array. If you've ever used conditional formatting in Excel, this is a simlar thing.

````python
df = pd.read_csv("schoolimprovement2010grants.csv")

# Create 2D array
pd_crosstab = pd.crosstab(df["Region"], df["Model Selected"])
print(pd_crosstab)
````

````python
# Plot a heatmap of the table
sns.heatmap(pd_crosstab)

plt.xticks(rotation=45)

plt.show()
````
