We now have a good idea on how to plot the basic Seaborn graphs. Seaborn graphs can be customised to look the way you want them. 

A lot of the customisation used in matplotlib can be used in Seaborn - since Seaborn is built off matplotlib - but there are some additional tricks that you should know

We will cover these now.

````python
# Initial Setup
%matplotlib inline

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

### Seaborn Styles
Start off with a simple example of customisation: setting the theme. There are five preset seaborn themes: ```darkgrid```, ```whitegrid```, ```dark```, ```white```, and ```ticks```. They are each suited to different applications and personal preferences. The default theme is darkgrid.

````python
df = pd.read_csv('FY18_4050_FMRs.csv')

plt.figure(figsize=(15, 5))

# Plot with a darkgrid style 
sns.set_style('darkgrid')
plt.subplot(1, 5, 1)
sns.distplot(df['fmr_2'])

# Plot with a whitegrid style 
sns.set_style('whitegrid')
plt.subplot(1, 5, 2)
sns.distplot(df['fmr_2'])

# Plot with a dark style 
sns.set_style('dark')
plt.subplot(1, 5, 3)
sns.distplot(df['fmr_2'])

# Plot with a white style 
sns.set_style('white')
plt.subplot(1, 5, 4)
sns.distplot(df['fmr_2'])

# Plot with a white style 
sns.set_style('ticks')
plt.subplot(1, 5, 5)
sns.distplot(df['fmr_2'])

plt.tight_layout()
plt.show()
````

### Colours
We saw in the previous notebooks how to set colours for matplotlib - as expected, these are available in seaborn. Additionally, Seaborn includes it's own colour methods using colour palettes: https://seaborn.pydata.org/tutorial/color_palettes.html

````python
# Create a magenta distplot
sns.distplot(df['fmr_3'], color='m')

# Show the plot
plt.show()
````

````python
plot_ind = 1
plt.figure(figsize=(15, 5))

for p in ['bright', 'colorblind', 'deep', 'muted', 'pastel']:
    sns.set_palette(p)
    plt.subplot(1, 5, plot_ind)
    sns.distplot(df['fmr_3'])
    plot_ind += 1
    
plt.show()
````

### Creating a Palette
Choosing a cohesive palette that works for your data can be time consuming. Fortunately, Seaborn has several functions that allow you to create your own custom sequential, categorical, or diverging palettes. Seaborn also makes it easy to see your palettes by using the ```palplot()``` function.

````python
# Create the coolwarm palette
sns.palplot(sns.color_palette("Purples", 8))
plt.show()

# Create the husl palette
sns.palplot(sns.color_palette("husl", 10))
plt.show()

# Create the coolwarm palette
sns.palplot(sns.color_palette("coolwarm", 6))
plt.show()
````

### Using Matplotlib
Seaborn uses ```matplotlib``` as the underlying library for creating plots. Most of the time, you can use the Seaborn API to modify your visualizations but sometimes it is helpful to use ```matplotlib```'s functions to customize your plots. The most important object in this case is ```matplotlib```'s axes.

We have in the past edited the main ```matplotlib``` class ```plt```. It is better to create axis and figure objects, and custom the individual objects.

````python
# Create a figure and axes
fig, ax = plt.subplots()

# Plot the distribution of 1 bedroom rents, specify the axes object
sns.distplot(df['fmr_1'], ax=ax)

# Modify the properties of the plot by setting the values of the axis object
ax.set(xlabel="1 Bedroom Fair Market Rent",
       xlim=(100,1500),
       title="US Rent")

# Calculate the median and mean
median = np.median(df['fmr_1'])
mean = np.mean(df['fmr_1'])

# Add vertical lines for the median and mean (another method of the axis object)
ax.axvline(x=median, color='m', label='Median', linestyle='--', linewidth=2)
ax.axvline(x=mean, color='b', label='Mean', linestyle='-', linewidth=2)

# Show the legend
ax.legend()
# Show the plot
plt.show()
````

### Multiple Plots
Last, we will briefly see how 2 plots are achieved using the axis object

````python
# Create a plot with 1 row and 2 columns that share the y axis label
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(10, 5))

# Plot the distribution of 1 bedroom apartments on ax0
sns.distplot(df['fmr_1'], ax=ax0)
ax0.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500))

# Plot the distribution of 2 bedroom apartments on ax1
sns.distplot(df['fmr_2'], ax=ax1)
ax1.set(xlabel="2 Bedroom Fair Market Rent", xlim=(100,1500))

# Display the plot
plt.show()
````
